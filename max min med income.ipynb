```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('DoctorVisits-DA.csv')

# Check the data types to ensure 'income' is numeric
df['income'] = pd.to_numeric(df['income'], errors='coerce')

# Display basic statistics for the 'income' column
print(df['income'].describe())
```

```python
plt.style.use('seaborn-v0_8')

# Create figure with three subplots
fig, (ax1) = plt.subplots(figsize=(7,6))

# 1. Histogram
sns.histplot(data=df, x='income', bins=30, ax=ax1)
ax1.set_title('Income Distribution')
ax1.set_xlabel('Income')
```

```python
plt.style.use('seaborn-v0_8')

# Create figure with three subplots
fig, (ax2) = plt.subplots(figsize=(7, 6))


# 2. Box plot
sns.boxplot(y=df['income'], ax=ax2)
ax2.set_title('Income Box Plot')
ax2.set_ylabel('Income')
```

```python
plt.style.use('seaborn-v0_8')

# Create figure with three subplots
fig, (ax3) = plt.subplots(figsize=(7, 6))


# 3. Violin plot
sns.violinplot(y=df['income'], ax=ax3)
ax3.set_title('Income Violin Plot')
ax3.set_ylabel('Income')
```

```python
print("Key statistics:")
print("Minimum income:", df['income'].min())
print("Maximum income:", df['income'].max())
print("Median income:", df['income'].median())
```
